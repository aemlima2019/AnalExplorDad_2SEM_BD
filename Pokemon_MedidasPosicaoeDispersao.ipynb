```python
print(base['Attack'].quantile(q=0.75)) # Calcula o Q3 (terceiro quartil ou percentil 75) da variável Attack
```

```python
# Calcula o desvio médio da variável Attack
def media(var):
    return sum(var)/len(var)

def desvio_medio(var):
    return media([abs(i - media(var)) for i in var])

print(desvio_medio(base['Attack']))
```

```python
print(base['Attack'].std()) # Calcula o desvio padrão da variável Attack
```

```python
x=base['Attack'].std() # Calcula a variância da variável Attack
y=x**2
print(y)

print((base['Attack'].std())**2) # Calcula a variância da variável Attack direto
```

```python
q1=base['Attack'].quantile(q=0.25) # Calcula a amplitude interquartil da variável Attack
q3=base['Attack'].quantile(q=0.75)
interquartil=q3-q1
print(interquartil)

print(base['Attack'].quantile(q=0.75)-base['Attack'].quantile(q=0.25))# Calcula a amplitude interquartil da variável Attack direto
```

```python
dp=base['Attack'].std() # Calcula o coeficiente de variação da variável Attack
media=base['Attack'].mean()
cv=(dp/media)*100
print(cv)

print((base['Attack'].std()/base['Attack'].mean())*100)# Calcula o coeficiente de variação da variável Attack direto
```
